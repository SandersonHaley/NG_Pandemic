```python
import pandas as pd
from scipy.stats import fisher_exact
from itertools import combinations
import numpy as np


# Assuming your DataFrame has columns named 'MLST' and 'Year'
# Filter the DataFrame for the specific MLST you are interested in
res_of_interest = 'R'  # Replace with the actual MLST you are interested in
MIC_PEN['Is_res_of_Interest'] = MIC_PEN['Penicillin'] == res_of_interest

# Group by year and calculate the counts of MLST of interest and total counts
yearly_counts = MIC_PEN[MIC_PEN['Year'].between(2017, 2022)].groupby('Year').agg(
    total_count=('Is_res_of_Interest', 'size'),
    res_count=('Is_res_of_Interest', 'sum')
).reset_index()

# Calculate the percentage of isolates with the specific MLST per year
yearly_counts['res_percentage'] = (yearly_counts['res_count'] / yearly_counts['total_count']) * 100

# Display the result
print(yearly_counts)

# Perform pairwise Fisher's exact tests with Bonferroni correction
years = yearly_counts['Year'].values
p_values = []

for (year1, year2) in combinations(years, 2):
    count1 = yearly_counts[yearly_counts['Year'] == year1][['res_count', 'total_count']].values[0]
    count2 = yearly_counts[yearly_counts['Year'] == year2][['res_count', 'total_count']].values[0]
    
    contingency_table = np.array([
        [count1[0], count1[1] - count1[0]],
        [count2[0], count2[1] - count2[0]]
    ])
    
    _, p = fisher_exact(contingency_table)
    p_values.append((year1, year2, p))

# Apply Bonferroni correction
alpha = 0.05
corrected_alpha = alpha / len(p_values)
significant_differences = [(year1, year2, p) for (year1, year2, p) in p_values if p < corrected_alpha]

# Display the results
print("\nPairwise Fisher's exact test p-values (with Bonferroni correction):")
for year1, year2, p in p_values:
    print(f"{year1} vs {year2}: p-value = {p}")

print(f"\nSignificant differences (p < {corrected_alpha}):")
for year1, year2, p in significant_differences:
    print(f"{year1} vs {year2}: p-value = {p}")
```

```python


import pandas as pd
from scipy.stats import fisher_exact
from itertools import combinations
import numpy as np


# Assuming your DataFrame has columns named 'MLST' and 'Year'
# Filter the DataFrame for the specific MLST you are interested in
mlst_of_interest = '11172'  # Replace with the actual MLST you are interested in
MLST['Is_MLST_of_Interest'] = MLST['ST'] == mlst_of_interest

# Group by year and calculate the counts of MLST of interest and total counts
yearly_counts = MLST[MLST['Year'].between(2017, 2022)].groupby('Year').agg(
    total_count=('Is_MLST_of_Interest', 'size'),
    mlst_count=('Is_MLST_of_Interest', 'sum')
).reset_index()

# Calculate the percentage of isolates with the specific MLST per year
yearly_counts['mlst_percentage'] = (yearly_counts['mlst_count'] / yearly_counts['total_count']) * 100

# Display the result
print(yearly_counts)

# Perform pairwise Fisher's exact tests with Bonferroni correction
years = yearly_counts['Year'].values
p_values = []

for (year1, year2) in combinations(years, 2):
    count1 = yearly_counts[yearly_counts['Year'] == year1][['mlst_count', 'total_count']].values[0]
    count2 = yearly_counts[yearly_counts['Year'] == year2][['mlst_count', 'total_count']].values[0]
    
    contingency_table = np.array([
        [count1[0], count1[1] - count1[0]],
        [count2[0], count2[1] - count2[0]]
    ])
    
    _, p = fisher_exact(contingency_table)
    p_values.append((year1, year2, p))

# Apply Bonferroni correction
alpha = 0.05
corrected_alpha = alpha / len(p_values)
significant_differences = [(year1, year2, p) for (year1, year2, p) in p_values if p < corrected_alpha]

# Display the results
print("\nPairwise Fisher's exact test p-values (with Bonferroni correction):")
for year1, year2, p in p_values:
    print(f"{year1} vs {year2}: p-value = {p}")

print(f"\nSignificant differences (p < {corrected_alpha}):")
for year1, year2, p in significant_differences:
    print(f"{year1} vs {year2}: p-value = {p}")
```
