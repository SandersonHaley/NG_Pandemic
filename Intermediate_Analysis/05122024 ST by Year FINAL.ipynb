```python
MAST_counts = yearst.groupby(["Year", "Final_NG-MAST"]).size().reset_index(name="Count")
```

```python
STAR_counts = yearst.groupby(["Year", "Final_NG-STAR"]).size().reset_index(name="Count")
```
