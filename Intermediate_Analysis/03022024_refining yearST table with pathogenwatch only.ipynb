```python
import os
import pandas as pd
```

```python
yearST = pd.read_csv("02162024_NG_yearST.csv", sep="\t")
```

```python
yearST
```

```python
def extract_second_number(x):
    if '/' in x:
        return x.split('/')[1]
    else:
        return x

# Applying the function to the column


yearST['PW_NG-STAR'] = yearST['NG-STAR'].apply(extract_second_number)
yearST['PW_NG-MAST'] = yearST['NG-MAST'].apply(extract_second_number)
```
