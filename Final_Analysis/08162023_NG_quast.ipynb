```python
#find all the quast results for HLF genomes and aggregate into one dataframe
import pandas as pd
from pathlib import Path
all_quast_results_set = []

quast_results_directory = "/media/haley/My Book1/Gonorrhea_Jo-Anne/08162023_NG_NCBI_QC_contigs/08042023_NG_quast/"
for quast_output in Path(quast_results_directory).glob("*_quast_results/transposed_report.tsv"):

    isolate_name = quast_output.parts[7]

    print(f"Parsing {quast_output}")
    isolate_quast_results = pd.read_csv(quast_output, sep='\t')
    isolate_quast_results['Isolate_Pathname'] = isolate_name

    all_quast_results_set.append(isolate_quast_results)

all_quast_results_set = pd.concat(all_quast_results_set)

print(all_quast_results_set)
```

```python
all_quast_results_set.to_csv("08162023_NG_Quast_results.csv")
```
