```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import fisher_exact
from statsmodels.stats.multitest import multipletests

# Load your data into a pandas DataFrame
# data = pd.read_csv('your_data.csv')

# Specify the sequence type you want to compare
target_sequence_type = '5985'

# Filter data for the specified sequence type
sequence_type_data = yearst[yearst['NG-MAST'] == target_sequence_type]

# Create a DataFrame to count occurrences of the sequence type in each year
sequence_type_counts = sequence_type_data['Year'].value_counts().sort_index()

# Calculate Fisher's exact test p-values for each year
p_values = []
for i in range(len(sequence_type_counts) - 1):
    year1 = sequence_type_counts.index[i]
    year2 = sequence_type_counts.index[i + 1]
    obs_table = pd.crosstab(sequence_type_data['Year'] == year1, sequence_type_data['Year'] == year2)
    _, p_value = fisher_exact(obs_table)
    p_values.append(p_value)

# Apply Bonferroni correction to p-values
adjusted_p_values = multipletests(p_values, method='bonferroni')[1]

# Create a bar plot with error bars representing p-values
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
bars = plt.bar(sequence_type_counts.index[:-1], sequence_type_counts.values[:-1], color='blue')  # Exclude the last year
plt.xlabel('Year')
plt.ylabel('Count')
plt.title(f'Count of NG-MAST Sequence Type {target_sequence_type} Across Years')

# Convert years to strings before setting the x-axis labels
years_as_strings = [str(year) for year in sequence_type_counts.index[:-1]]  # Exclude the last year
plt.xticks(sequence_type_counts.index[:-1], years_as_strings, rotation=45)

# Add error bars representing adjusted p-values
for i, bar in enumerate(bars):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 2, f'p={adjusted_p_values[i]:.3f}', ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```
